```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew','popularity']]
```

```python
popular_movies=movies.sort_values("popularity",ascending=False)
```

```python
popular_movies.head()
```

```python
new = popular_movies.drop(columns=['overview','genres','keywords','cast','crew'])
```
